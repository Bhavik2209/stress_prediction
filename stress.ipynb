```python
df['SBP'] = ''
df['DBP'] = ''
```

```python
for i in range(len(df)):
    k = df['Blood Pressure'][i].split("/")[0]
    j = df['Blood Pressure'][i].split("/")[1]
    df['SBP'][i] = k
    df['DBP'][i] = j
```

```python
df.drop(columns='Blood Pressure',inplace=True)
```

```python
one_hot_encoded = pd.get_dummies(df['Sleep Disorder'], prefix='Sleep_Disorder',dtype=int)
df= pd.concat([df, one_hot_encoded], axis=1)
```

```python
df['Sleep_Disorder_Sleep Apnea']
```

```python
df.drop(columns='Sleep Disorder',inplace=True)
```

```python
X = df.drop(columns='Stress Level')
y = df['Stress Level']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test  = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
from lazypredict.Supervised import LazyClassifier
clf = LazyClassifier(verbose=0,ignore_warnings=True,custom_metric=None)
```

```python
models,predictions = clf.fit(X_train,X_test,y_train,y_test)
print(models)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
# Save scaler and model using pickle
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import pickle
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)
```
